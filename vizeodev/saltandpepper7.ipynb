```python
import random
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('gamma.jpg', cv2.IMREAD_GRAYSCALE)
histogram, bin_edges = np.histogram(img, bins=256, range=(0, 1))
def add_noise(img):
    row, col = img.shape
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        y_coord = random.randint(0, row - 1)
        x_coord = random.randint(0, col - 1)
        img[y_coord][x_coord] = 255
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        y_coord = random.randint(0, row - 1)
        x_coord = random.randint(0, col - 1)
        img[y_coord][x_coord] = 0
    return img
plt.figure(figsize=(8, 8))
plt.subplot(2, 2, 1), plt.plot(bin_edges[0:-1], histogram)
plt.subplot(2, 2, 2), plt.imshow(img, cmap='gray')
plt.title("Without Noise")

img1 = add_noise(img)
histogram1, bin_edges1 = np.histogram(img1, bins=256, range=(0, 1))

plt.subplot(2, 2, 3), plt.plot(bin_edges1[0:-1], histogram1)
plt.subplot(2, 2, 4), plt.imshow(img1, cmap='gray')
plt.title("With Noise")
plt.show()
```
